```python
from sklearn.utils import resample
```

```python
fully_paid = df[df["Loan Status"]=="Fully Paid"]
charged_off = df[df["Loan Status"]=="Charged Off"]
```

```python
downsampled_data = resample(fully_paid, replace=False, n_samples=len(charged_off))
```

```python
downsampled_data
```

```python
charged_off
```

```python
new_df = pd.concat([downsampled_data, charged_off], axis=0)
```

```python
new_df.reset_index(drop=True, inplace=True)
```

```python
new_df.info()
```

```python
X = new_df.drop("Loan Status", axis=1)
y = new_df["Loan Status"]
```

```python
y.shape
```

```python
# PCA implementation
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
new_X = pca.fit_transform(X)
new_X
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
X_tr.shape
```

```python
X_te.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_tr = scaler.fit_transform(X_tr)
```

```python
X_te = scaler.transform(X_te)
```

```python
svc = SVC(kernel="linear")
```

```python
svc.fit(X_tr, y_tr)
```

```python
y_pr = svc.predict(X_te)
```

```python
accuracy_score(y_te, y_pr)
```

```python
svc.score(X_tr, y_tr)
```

```python
model = SVC(C=5)
```

```python
model.fit(X_tr, y_tr)
```

```python
model.score(X_tr, y_tr)
```

```python
y_pr2 = model.predict(X_te)
```

```python
accuracy_score(y_te, y_pr2)
```

```python
print(classification_report(y_te, y_pr2))
```

```python
confusion_matrix(y_te, y_pr2)
```

```python
m2 = SVC(kernel="poly",C=5)
```

```python
m2.fit(X_tr, y_tr)
print("Training Score",m2.score(X_tr, y_tr))
print("Testing Score", accuracy_score(y_te, m2.predict(X_te)))
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
base_estimators = [("svc", SVC(kernel="rbf")),
                   ("dt", DecisionTreeClassifier(criterion="entropy", max_depth=5)),
                    ("logreg", LogisticRegression())]
```

```python
model1 = DecisionTreeClassifier(criterion="entropy", max_depth=5)
```

```python
sc = StackingClassifier(estimators=base_estimators, final_estimator=model1, cv=5,
                       n_jobs=-1, verbose=1)
```

```python
sc.fit(X_tr, y_tr)
```

```python
sc.predict(X_te)
```

```python
sc.score(X_tr, y_tr)
```

```python
y_pr3 = sc.predict(X_te)
```

```python
accuracy_score(y_te, y_pr3)
```
