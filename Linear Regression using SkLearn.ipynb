```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.datasets import make_regression
```

```python
# Generating data using make_regression

data = make_regression(n_samples=1000, n_features=1, noise=25)
```

```python
X = data[0]
y = data[1]
```

```python
df = pd.DataFrame(X, columns=["Feature1"])
df["Target"] = y
```

```python
plt.title("Feature vs Target")
plt.scatter(df["Feature1"], df["Target"])
plt.show()
```

```python
# Train test split

# Return ----------->
# X_train -> All the input features for training
# X_test -> All the input features for testing
# y_train -> Target column for training
# y_test -> Target for testing

# Input ----------->
# X -> All the input features from numpy array or pandas dataframe
# y -> Target variable as numpy array or pandas dataframe column
# train_size or test_size (0.0 to 1.0)-> To determine how many rows will be in train data 
# and how many rows will be in test data
# random_state -> To ensure the data not to change at every refresh
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=57)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_mean = y_test.mean()
y_mean
```

```python
TSS = np.sum(np.square(y_test - y_mean))
TSS
```

```python
SSR = np.sum(np.square(y_test- y_pred))
SSR
```

```python
r2 = 1 - (SSR / TSS)
r2
```
