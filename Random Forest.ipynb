```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data = make_classification(n_samples=1000, n_classes=3, n_informative=6)
```

```python
X = data[0]
y = data[1]
```

```python
X[:10]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[y==0,0], X[y==0,2])
plt.scatter(X[y==1,0], X[y==1,2])
plt.scatter(X[y==2,0], X[y==2,2])
```

```python
model = RandomForestClassifier(n_estimators=100, n_jobs=-1, criterion="entropy")
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, y, test_size=.25, random_state=43)
```

```python
model.fit(X_tr, y_tr)
```

```python
model.score(X_tr, y_tr)
```

```python
y_pr = model.predict(X_te)
```

```python
accuracy_score(y_te, y_pr)
```

```python
conf = confusion_matrix(y_te, y_pr)
conf
```

```python
print(classification_report(y_te, y_pr))
```

```python
import seaborn as sns
```

```python
sns.heatmap(conf, annot=True)
```
