```python
lr = 0.01
b = -19
m = 90
epochs = 15
for i in range(epochs):
    
    b_slope = -2 * np.sum(out - m * inp.ravel() - b)
    m_slope = -2 * np.sum((out - m * inp.ravel() - b)* inp.ravel())
    
    b = b - (lr * b_slope)
    m = m - (lr * m_slope)

    temp = m*inp + b
    plt.scatter(inp, out)
    plt.plot(inp, y_pred, color="red")
    plt.plot(inp, temp, color="blue")
plt.show()
```
