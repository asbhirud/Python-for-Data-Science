```python
from sklearn.datasets import make_regression
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = make_regression(n_samples = 6, n_features = 1, n_targets = 1, noise = 12, random_state = 243)
```

```python
data
```

```python
X = data[0]
y = data[1]
```

```python
plt.scatter(X,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X, y)
```

```python
m = lr.coef_
m
```

```python
intercept = lr.intercept_
intercept
```

```python
y_pred = lr.predict(X)
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X, y_pred, "r-")
plt.show()
```
