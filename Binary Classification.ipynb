```python
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data = make_classification(n_samples=100, n_features=4, n_classes=2, weights=[0.5, 0.5])
```

```python
data
```

```python
type(data)
```

```python
X = data[0]
y = data[1]
```

```python
df = pd.DataFrame(X, columns=[f"inp{i+1}" for i in range(4)])
```

```python
df["target"] = y
```

```python
df.head()
```

```python
plt.scatter(df["inp1"][df["target"]==1], df["inp1"][df["target"]==0])
plt.scatter(df["inp4"][df["target"]==1], df["inp4"][df["target"]==0])
```

```python
df.drop(["inp2", "inp3"], axis=1, inplace=True)
```

```python
df.head()
```

```python
model = LogisticRegression()
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
model.fit(X, y)
```

```python
model.score(X, y) # Training Score
```

```python
y_pred = model.predict(X)
```

```python
accuracy_score(y, y_pred) # Testing Accuracy
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y, y_pred)
```
