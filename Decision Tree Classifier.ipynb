```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["target"] = iris.target
```

```python
df.head(3)
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=234)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(15,15))
tree.plot_tree(model)
plt.show()
```

```python
model2 = DecisionTreeClassifier(criterion="entropy")
```

```python
model2.fit(X_train, y_train)
```

```python
plt.figure(figsize=(15,15))
tree.plot_tree(model2)
plt.show()
```

```python
model3 = DecisionTreeClassifier(criterion="entropy", max_depth=2)
```

```python
model3.fit(X_train, y_train)
```

```python
y_pred3 = model3.predict(X_test)
```

```python
accuracy_score(y_test, y_pred3)
```

```python
print(classification_report(y_test, y_pred3))
```

```python
model4 = DecisionTreeClassifier(criterion="entropy", max_depth=2, max_features=2)
```

```python
model4.fit(X_train, y_train)
```

```python
y_pred4 = model4.predict(X_test)
```

```python
accuracy_score(y_test, y_pred4)
```

```python
plt.figure(figsize=(6,6))
tree.plot_tree(model4)
plt.show()
```
